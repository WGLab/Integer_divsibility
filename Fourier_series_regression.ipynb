```python
import numpy as np
import pandas as pd
import random
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
import statsmodels.api as sm
```

## Mod 3

```python
l1 = range(2**32-1)
X = random.sample(l1,200)
y = np.mod(X,3)
train_df = pd.DataFrame(zip(X,y), columns=['int','mod3'])
train_df['sin1'] = np.sin(2 * np.pi * 1 * train_df['int']/3)
train_df['cos1'] = np.cos(2 * np.pi * 1 * train_df['int']/3)
train_df['sin2'] = np.sin(2 * np.pi * 2 * train_df['int']/3)
train_df['cos2'] = np.cos(2 * np.pi * 2 * train_df['int']/3)
train_df.head(5)
```

```python
X = train_df[['sin1','cos1']].iloc[:150]
y = train_df['mod3'].iloc[:150]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
```

```python
LM = LinearRegression()
LM.fit(X_train,y_train)
print(LM.score(X_test, y_test)) #R^2 value
print(LM.intercept_)
```

```python
pd.DataFrame(zip(X.columns, LM.coef_)).set_axis(['Feature','Coefficient'], axis='columns')
```

```python
X_val = train_df[['sin1','cos1']].iloc[-50:]
y_val = train_df['mod3'].iloc[-50:]
y_val = y_val.to_list()
y_val_predict = LM.predict(X_val)
```

```python
error = 10 **(-5)
y_val_pred_label = [0 if abs(y-0)< error else 1 if abs(y-1) < error else 2 for y in y_val_predict]
```

```python
count = 0
for i in range(len(y_val_pred_label)):
  if y_val[i] == y_val_pred_label[i]:
    count += 1
accuracy = count/len(y_val_pred_label)
print(accuracy)
```

## Mod  7

```python
l1 = range(2**32-1)
X = random.sample(l1,200)
y = np.mod(X,7)
train_df = pd.DataFrame(zip(X,y), columns=['int','mod7'])
train_df['sin1'] = np.sin(2 * np.pi * 1 * train_df['int']/7)
train_df['cos1'] = np.cos(2 * np.pi * 1 * train_df['int']/7)
train_df['sin2'] = np.sin(2 * np.pi * 2 * train_df['int']/7)
train_df['cos2'] = np.cos(2 * np.pi * 2 * train_df['int']/7)
train_df['sin3'] = np.sin(2 * np.pi * 3 * train_df['int']/7)
train_df['cos3'] = np.cos(2 * np.pi * 3 * train_df['int']/7)
train_df['sin4'] = np.sin(2 * np.pi * 4 * train_df['int']/7)
train_df['cos4'] = np.cos(2 * np.pi * 4 * train_df['int']/7)
train_df['sin5'] = np.sin(2 * np.pi * 5 * train_df['int']/7)
train_df['cos5'] = np.cos(2 * np.pi * 5 * train_df['int']/7)
train_df['sin6'] = np.sin(2 * np.pi * 6 * train_df['int']/7)
train_df['cos6'] = np.cos(2 * np.pi * 6 * train_df['int']/7)
train_df.head(5)
```

```python
X7 = train_df[['sin1','cos1','sin2','cos2','sin3','cos3']].iloc[:150]
y7 = train_df['mod7'].iloc[:150]
X7_train, X7_test, y7_train, y7_test = train_test_split(X7, y7, test_size = 0.1)
```

```python
LM7 = LinearRegression()
LM7.fit(X7_train,y7_train)
print(LM7.score(X7_test, y7_test)) #R^2 value
print(LM7.intercept_)
```

```python
pd.DataFrame(zip(X7.columns, LM7.coef_)).set_axis(['Feature','Coefficient'], axis='columns')
```

```python
X7_val = train_df[['sin1','cos1','sin2','cos2','sin3','cos3']].iloc[-50:]
y7_val = train_df['mod7'].iloc[-50:]
y7_val = y7_val.to_list()
y7_val_predict = LM7.predict(X7_val)
```

```python
error = 10 **(-5)
y7_val_pred_label = [0 if abs(y-0)< error else 1 if abs(y-1) < error else 2 if abs(y-2) < error else 3 if abs(y-3) < error else 4 if abs(y-4) < error else 5 if abs(y-5) < error else 6 for y in y7_val_predict]
count7 = 0
for i in range(len(y7_val_pred_label)):
  if y7_val[i] == y7_val_pred_label[i]:
    count7 += 1
accuracy7 = count7/len(y7_val_pred_label)
print(accuracy7)
```
